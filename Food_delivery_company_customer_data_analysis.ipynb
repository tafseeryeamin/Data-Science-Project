```python
#percentage of zero rating
total_orders_by_country = final_df.groupby('Country').size().reset_index(name='Total Orders')
zero_ratings_by_country = final_df[final_df['Rating color'] == 'White'].groupby('Country').size().reset_index(name='Zero Rating Count')
percentage=zero_ratings_by_country.merge(total_orders_by_country, on='Country', how='left')
percentage['Percentage'] = (percentage['Zero Rating Count'] / percentage['Total Orders']) * 100
percentage
```

```python
# Step 1: Calculate the total number of orders placed by each country
total_orders_by_country = final_df.groupby('Country').size().reset_index(name='Total Orders')

# Step 2: Filter rows with poor ratings ('Red' or 'Orange') and calculate their count by country
poor_rating_by_country = final_df[final_df['Rating color'].isin(['Red', 'Orange'])].groupby('Country').size().reset_index(name='Not satisfied Rating Count')

# Step 3: Merge the two DataFrames on 'Country'
percentage = poor_rating_by_country.merge(total_orders_by_country, on='Country', how='right')

# Step 4: Handle missing values (countries with no poor ratings)
percentage['Not satisfied Rating Count'] = percentage['Not satisfied Rating Count'].fillna(0)

# Step 5: Calculate the percentage of poor ratings
percentage['Percentage'] = (percentage['Not satisfied Rating Count'] / percentage['Total Orders']) * 100

# Step 6: Sort the DataFrame by 'Percentage' in descending order (optional)
percentage = percentage.sort_values(by='Percentage', ascending=False)

# Display the result
percentage
```

```python
# Access multiple columns including 'City'
selected_columns = final_df[['Average Cost for two', 'City','Country']]

# Group by 'City' and calculate the mean of 'Average Cost for two'
grouped_data = selected_columns.groupby('City')['Average Cost for two'].mean().reset_index()

# Display the resulting DataFrame
grouped_data.head(50)
```

```python
# Step 1: Filter rows where 'Country' is 'India'
filtered_data = final_df[final_df['Country'] == 'India']

# Step 2: Select the required columns
selected_columns = filtered_data[['Average Cost for two', 'City']]

# Step 3: Group by 'City', calculate the mean of 'Average Cost for two', and count the number of orders
grouped_data = selected_columns.groupby('City').agg(
    Average_Cost=('Average Cost for two', 'mean'),  # Calculate mean
    Number_of_Orders=('Average Cost for two', 'size')  # Count number of orders
).reset_index()
grouped_data = grouped_data.sort_values(by='Number_of_Orders', ascending=False)
# Step 4: Display the resulting DataFrame
grouped_data
```

```python
import matplotlib.pyplot as plt
import numpy as np



# Create the plot
fig, ax1 = plt.subplots(figsize=(50, 6))

# Bar plot for Average_Cost
ax1.bar(
    grouped_data['City'],
    grouped_data['Average_Cost'],
    color='blue',
    alpha=0.6,
    label='Average Cost'
)
ax1.set_xlabel('City', fontsize=12)
ax1.set_ylabel('Average Cost for Two', fontsize=12)
ax1.tick_params(axis='y')

# Line plot for Number_of_Orders
ax2 = ax1.twinx()
ax2.plot(
    grouped_data['City'],
    grouped_data['Number_of_Orders'],
    color='red',
    marker='o',
    label='Number of Orders'
)
ax2.set_ylabel('Number of Orders', fontsize=12)
ax2.tick_params(axis='y')

# Add title and legends
plt.title('Average Cost and Number of Orders by City', fontsize=16)
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))

# Show the plot
plt.show()
```

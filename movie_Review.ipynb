```python
# @title release_date vs popularity

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['release_date']
  ys = series['popularity']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(20, 5.2), layout='constrained')
df_sorted = df.sort_values('release_date', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('original_language')):
  _plot_series(series, series_name, i)
  fig.legend(title='original_language', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('release_date')
_ = plt.ylabel('popularity')
```

```python
# @title release_date vs vote_count


def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['release_date']
  ys = series['vote_count']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(15, 5.2), layout='constrained')
df_sorted = df.sort_values('release_date', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('original_language')):
  _plot_series(series, series_name, i)
  fig.legend(title='original_language', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('release_date')
_ = plt.ylabel('vote_count')
```

```python

C = df['vote_average'].mean()

# Determine a minimum vote count threshold (m) – for example, the 80th percentile of vote counts
m = df['vote_count'].quantile(0.50)

# Filter out movies that qualify (with vote_count greater than or equal to m)
qualified_movies = df[df['vote_count'] >= m]

qualified_movies
```

```python
# Define the weighted rating function
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v / (v + m) * R) + (m / (v + m) * C)

# Apply the weighted rating function to the qualified movies
qualified_movies['weighted_rating'] = qualified_movies.apply(weighted_rating, axis=1)

# Sort by the weighted rating and get the top 10
t=qualified_movies.sort_values('weighted_rating', ascending=False).head(10)
a=t.reset_index()
# Display the top 10 movies
print(a[['title', 'weighted_rating', 'vote_average', 'vote_count']])
```
